```python
data = {'planet': ['Mercury', 'Venus', 'Earth', 'Mars'], 
        'radius_km': [2440, 6052, 6371, 3390], 
        'moons': [0, 0, 1, 2],
}

df1 = pd.DataFrame(data)
df1
```

```python
data = {'planet': ['Jupiter', 'Saturn', 'Uranus', 'Neptune'], 
        'radius_km': [69911, 58232, 25362, 24622], 
        'moons': [80, 83, 27, 14],
}

df2 = pd.DataFrame(data)
df2
```

```python
df3 = pd.concat([df1, df2], axis = 0)
df3
```

```python
df3 = df3.reset_index(drop=True)
df3
```

```python
data = {
    "planet": [
        "Earth", "Mars", "Jupiter", "Saturn",
        "Uranus", "Neptune", "Janssen", "Tadmor"
    ],
    "type": [
        "terrestrial", "terrestrial", "gas giant", "gas giant",
        "ice giant", "ice giant", "super earth", "gas giant"
    ],
    "rings": [
        "no", "no", "yes", "yes", "yes", "yes", "no", None
    ],
    "mean_temp_c": [
        15.0, -65.0, -110.0, -140.0, -195.0, -200.0, np.nan, np.nan
    ],
    "magnetic_field": [
        "yes", "no", "yes", "yes", "yes", "yes", None, "no"
    ],
    "life": [
        1, 0, 0, 0, 0, 0, 1, 1
    ]
}

df4 = pd.DataFrame(data)
df4
```

```python
inner = pd.merge(df3, df4, on="planet", how="inner")
inner
```

```python
outer = pd.merge(df3, df4, on="planet", how="outer")
outer
```

```python
left = pd.merge(df3, df4, on="planet", how="left")
left
```

```python
right = pd.merge(df3, df4, on="planet", how="right")
right
```
